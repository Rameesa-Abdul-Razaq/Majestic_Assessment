```python
import matplotlib.pyplot as plt

# Soft Drink Data (Litres)

drinks = ["Cola", "Diet Cola", "Lemonade", "Dandelion & Burdock", "Orangeade"]
litres = [15, 20, 25, 60, 5]

plt.figure()
plt.bar(drinks, litres, color="purple")
plt.title("Soft Drinks Consumed at Student Hackathon")
plt.xlabel("Drink Type")
plt.ylabel("Litres Consumed")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Sweets Data (grams)

sweets = ["Teacakes", "Lollipops", "Parma Violets", "Chocolate Coins"]
mass = [900, 700, 3000, 400]

plt.figure()
plt.bar(sweets, mass, color="purple")
plt.title("Mass of Sweets Consumed at Student Hackathon")
plt.xlabel("Sweet Type")
plt.ylabel("Mass (grams)")
plt.show()
```
